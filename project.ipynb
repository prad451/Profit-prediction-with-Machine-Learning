Pieplot

```python
plt.figure(figsize=(8,8))
labels = ["R&D Spend","Administration","Marketing Spend"]
size = [df["R&D Spend"].mean(),df["Administration"].mean(),df["Marketing Spend"].mean()]
Explode = [0.1,0.1,0.1]
plt.pie(size,explode=Explode,labels=labels,shadow=True,startangle=45,autopct="%1.1f%%")
plt.title("PiePlot")
plt.legend(title="Spend")
plt.axis("equal")
plt.show()
```

**Splitting train and test dataset**

```python
X = df.iloc[:,0:3]
Y = df.iloc[:,4]
```

```python
X
```

```python
Y
```

**Model Building**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
Y_pred
```

```python
result = pd.DataFrame({'Predicted profit ': Y_pred.flatten()})
```
